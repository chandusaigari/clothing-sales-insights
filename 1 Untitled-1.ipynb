```python
import pandas as pd

# Function to convert AgeGroup to integer
def convert_age_group(x):
    if pd.isna(x):
        return pd.NA               # handle missing values
    x = str(x).strip()             # ensure string type and remove spaces
    if "+" in x:                   # handle "36+" or "56+"
        return int(x.replace("+", ""))
    elif "-" in x:                 # handle ranges "15-25"
        low, high = map(int, x.split("-"))
        return (low + high) // 2   # take midpoint
    else:
        return pd.NA

# Apply to your df and replace the column
df["AgeGroup"] = df["AgeGroup"].apply(convert_age_group).astype("Int64")  # nullable integer

# Verify
print(df.info())
print(df[["AgeGroup"]].head())
```
